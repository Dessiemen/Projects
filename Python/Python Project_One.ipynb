#### 2. Cleaning and Transforming the Data

Check whether the file contains null values.

```python

VGS.isnull().sum()
```

**Remove null valles from the VGS file.** 

```python
 

VGS = VGS.dropna()
```

Reset the index after removing null values.

```python
VGS1 = VGS.reset_index()
VGS1
```

Drop the old "index" column and "Rank" column for calculation purpose.

```python

VGS1 = VGS1.drop(columns=["index", "Rank"])
```

Display the VGS1 file after dropping some columns.

```python

VGS1
```

Convert Year data type from float data type to int data type.

```python

VGS1["Year"] = VGS1["Year"]. astype(int)
```

Check whether the 'Year' data type is converted.

```python

print('Year data type:', VGS1['Year'].dtype)
```

#### 3. Grouping, Aggregating, and Plotting

Group and aggregate the data for analysis purpose.

```python

VGS1 = VGS1.groupby("Year").sum()
```

Plot a line graph to know the trends of regions and global sales over the years.

```python

VGS1.plot(ylabel = "Sales", title = "Regions and Global Sales by Year")
```

Plot a bar graph to compare the regional sales against the years.

```python

VGS1.plot(kind = "bar", figsize = (12,5), ylabel = "Sales",title = "Regions sales by Year")
```

Plot a pie chart to compare regional sales over the years.

```python

VGS1[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']].sum().plot(kind ="pie", autopct = '%1.1f%%', cmap = "tab20c",
                 figsize = (8, 6), title = "Regions Sales from 1998-2020 " )
                                                            
```

Group and aggregate the data.

```python
VGS2 = VGS[['Genre','Global_Sales','NA_Sales','EU_Sales','JP_Sales','Other_Sales']].groupby('Genre').sum()
```

Plot a bar graph for global and regional sales by year.

```python

VGS2.sort_values(by = 'Global_Sales', ascending = False).plot(kind = "bar",figsize = (10, 6), width = 0.8, ylabel="Sales")
```

Plot a pie chart of global sales by genre.

```python
VGS2['Global_Sales'].groupby('Genre').sum().plot(kind ='pie', title = 'Global Sales by Genre', ylabel=' ',
                        figsize =(9,7), autopct = '%1.1f%%', cmap = "tab20c")
```
